```python
#Julia Gajtkowska - Zadanie 3
# import potrzebnych bibliotek
import matplotlib.pyplot as plt
import numpy as np 

# ilość punktów
N = 500

# zakresy wykresów na osi X
x1 = np.linspace(-1,1,N)
x2 = np.linspace(-5,5,N)
x3 = np.linspace(0,5,N)

def foo_plot(x):
  # stworzenie figury i jej wykresu
  fig = plt.figure() 
  ax = fig.add_subplot(1,1,1)
  fig.tight_layout()
  # dodanie tytułu, legend i opisów osi
  ax.set_title('Wykres funkcji y=x^2 + 5', loc='center') 
  ax.set_ylabel('Y') 
  ax.set_xlabel('X')
  line, = ax.plot(x, x**2 + 5,label='y=x^2 + 5')
  ax.legend(loc='best', frameon=False) 
```

```python
foo_plot(x1)
```

```python
foo_plot(x2)
```

```python
foo_plot(x3)
```
