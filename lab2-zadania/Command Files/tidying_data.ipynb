```python
import pandas as pd 
import numpy as np

# Na początek wyczyściłam dane z niepoptrzebnych i przeszkadzających elementów, następnie zapisałam wynik do pliku 
# weather_clean.csv

infile = "weather.txt"
outfile = "weather1.txt"

delete_list = [" I", " S", " SI", " IS"," OI", " SO", " OS", " IO", " I ", " S ","  ","   ", "    ","     ", "      ", "       ", "        "]
spaces = ["  ","   ", "    ","     ", "      ", "       ", "        "]

with open(infile) as fin, open(outfile, "w+") as fout:
    for line in fin:
        line = line.replace("-", " -")
        line = line.replace("-9999", 'Null')
        line = line.replace("9999", 'Null')
        for word in delete_list:
            line = line.replace(word, " ")
            line = line.replace(word, " ")
        for word in spaces:
            line = line.replace(word, " ")
        fout.write(line)
    

df = pd.read_csv("weather1.txt", sep=" ", header=None)
df = df.drop(columns = 32)
df = df.rename(columns={0: "Date"})

df.to_csv('weather_clean.csv')

df.head()
```

```python
# Dodanie kolumn ID, Year, Month, Variable

df.insert(1, "ID", "")
df.insert(2, "Year", "")
df.insert(3, "Month", "")
df.insert(4, "Variable", "")
```

```python
# Wypełnienie wszystkich kolumn i usunięcie kolumny "Date"
for e in range(0,len(df["Date"]),1):
    df.loc[e,"ID"] = df.loc[e,"Date"][:11]
    df.loc[e,"Year"] = df.loc[e,"Date"][11:15]
    df.loc[e,"Month"] = df.loc[e,"Date"][15:17]
    df.loc[e,"Variable"] = df.loc[e,"Date"][17:]

df = df.drop(["Date"],axis=1)

# Zmniejszenie wielkości tabeli (utworzenie kolumn variable i value)
    
df = pd.melt(df, id_vars=["ID", "Year", "Month", "Variable"])

# Usunięcie wierszy z zwartościami Null

df = df[df["value"] != "Null"]

# Dodanie kolumny z pełną datą i usunięcie kolumn Day i Month
df = df.rename(columns={"variable":"Day","Year":"Date"})

df["Date"] = df["Date"].astype(str) + '-' + df["Month"].astype(str) +'-'+ df["Day"].astype(str)

df = df.drop(["Month", "Day"], axis=1)

df.head()

# Zapisanie wyniku do pliku Weather_tidy.csv

df.to_csv('weather_tidy.csv')
```
